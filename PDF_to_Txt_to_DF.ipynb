Once `pypdf` is installed, you can use the following code to convert a PDF file to plain text. Remember to replace `'your_document.pdf'` with the actual path to your PDF file.

```python
import pypdf

def convert_pdf_to_text(pdf_path, output_text_path='output.txt'):
    """
    Converts a PDF file to plain text.

    Args:
        pdf_path (str): The path to the input PDF file.
        output_text_path (str): The path to save the extracted text.
    """
    try:
        reader = pypdf.PdfReader(pdf_path)
        text = ""
        for page in reader.pages:
            text += page.extract_text() + "\n"

        with open(output_text_path, 'w', encoding='utf-8') as f:
            f.write(text)
        print(f"Successfully extracted text to {output_text_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
# First, make sure you have a PDF file available, for instance, upload 'example.pdf' to your Colab environment.
# Then uncomment and call the function with the path to your PDF.
convert_pdf_to_text('/content/Zosia_Oceny.pdf', '/content/Zosia_Oceny.txt')

print("To run this code, please upload a PDF file (e.g., 'example.pdf') to your Colab environment.")
print("Then uncomment and run the 'convert_pdf_to_text' line with your file name.")
```

## Parse Text to DataFrame

### Subtask:
Parse the extracted text content into a pandas DataFrame, separating student information, subject names, and their corresponding grades.


**Reasoning**:
The subtask requires parsing the extracted text into a pandas DataFrame. I will follow the instructions to first extract student information, define column names, then iterate through the grade lines, use regular expressions to separate subject names and grade values, and finally construct a DataFrame, handling cases where grade values might be missing by padding with `None`.



```python
import pandas as pd
import re

# 2. Extract student's name and class from the 'lines' list
# 'lines' was generated in a previous step by splitting the 'content' string.
student_name_line = lines[1]
student_name = student_name_line.split(':', 1)[1].strip() if ':' in student_name_line else student_name_line.strip()

student_class_line = lines[2]
student_class = student_class_line.split(':', 1)[1].strip() if ':' in student_class_line else student_class_line.strip()

print(f"Student Name: {student_name}")
print(f"Student Class: {student_class}")

# 3. Define a list of column names for the DataFrame
columns = [
    'Przedmiot',
    'Oceny_Biezace_Okres1', 'Sr_I_Okres1', 'I_Okres1',
    'Oceny_Biezace_Okres2', 'Sr_II_Okres2', 'II_Okres2',
    'Sr_R_KoniecRoku', 'R_KoniecRoku'
]

# 4. Initialize an empty list to store dictionaries of parsed data
grade_data = []

# Regex to separate the subject name from the grade values
subject_grade_regex = re.compile(r'(.+?)\s+([\s\S]*)')

# Regex to capture individual grade tokens, including bracketed values and "Brak ocen"
# This regex looks for:
# 1. \[.*?\]: Anything enclosed in square brackets (non-greedy)
# 2. Brak\socen: The exact phrase "Brak ocen"
# 3. \S+: Any sequence of non-whitespace characters (for single grades like numbers, +, -, np, bz)
grade_token_regex = re.compile(r'(\[.*?\]|Brak\socen|\S+)')

# 5. Iterate through the lines of the 'content' starting from line 7 (index 6)
# This assumes each logical subject-grade entry starts on a new line from this point.
for line in lines[6:]:
    line = line.strip()
    if not line:  # Skip empty lines
        continue

    match = subject_grade_regex.match(line)
    if match:
        subject_name = match.group(1).strip()
        grades_string = match.group(2).strip()

        # Extract individual grade entries using the refined regex
        grade_tokens = grade_token_regex.findall(grades_string)

        # Ensure we have exactly 8 grade values, padding with None if fewer, truncating if more
        expected_grade_columns = len(columns) - 1 # 8 grade columns
        if len(grade_tokens) < expected_grade_columns:
            grade_tokens.extend([None] * (expected_grade_columns - len(grade_tokens)))
        elif len(grade_tokens) > expected_grade_columns:
            grade_tokens = grade_tokens[:expected_grade_columns] # Truncate if too many

        # Create a dictionary for the current subject, mapping to defined column names
        subject_dict = {
            'Przedmiot': subject_name,
            'Oceny_Biezace_Okres1': grade_tokens[0],
            'Sr_I_Okres1': grade_tokens[1],
            'I_Okres1': grade_tokens[2],
            'Oceny_Biezace_Okres2': grade_tokens[3],
            'Sr_II_Okres2': grade_tokens[4],
            'II_Okres2': grade_tokens[5],
            'Sr_R_KoniecRoku': grade_tokens[6],
            'R_KoniecRoku': grade_tokens[7]
        }
        grade_data.append(subject_dict)
    else:
        # Print lines that couldn't be parsed by the subject_grade_regex for inspection
        print(f"Warning: Could not parse line for subject/grades: {line}")

# 6. Convert the list of dictionaries into a pandas DataFrame
df = pd.DataFrame(grade_data, columns=columns)

# Display the DataFrame and its info to verify parsing
print("\nParsed DataFrame (first 5 rows):")
print(df.head())
print("\nDataFrame Info:")
df.info()
```
